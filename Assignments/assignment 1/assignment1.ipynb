### Task 1

```python
transactions = [
    {"id": 1, "product": "Laptop", "price": 1200, "quantity": 2, "region": "EU"},
    {"id": 2, "product": "Phone", "price": 800, "quantity": -1, "region": "US"},
    {"id": 3, "product": "Tablet", "price": 0, "quantity": 5, "region": "EU"},
    {"id": 4, "product": "Monitor", "price": 350, "quantity": 3, "region": "US"},
    {"id": 5, "product": "Laptop", "price": 1200, "quantity": 1, "region": "EU"},
    {"id": 6, "product": "Keyboard", "price": 75, "quantity": 10, "region": "APAC"},
]

# Filter out invalid transactions
valid_transactions = []
for t in transactions:
    if t["price"] != 0 and t["quantity"] >= 0:
        valid_transactions.append(t)

# Calculate total for each valid transaction
for t in valid_transactions:
    t["total"] = t["price"] * t["quantity"]

# Group transactions by region to find total revenue
revenue_per_region = {}
for t in valid_transactions:
    region = t["region"]
    
    if region not in revenue_per_region:
        revenue_per_region[region] = 0
        
    revenue_per_region[region] = revenue_per_region[region] + t["total"]

# Find the region with the maximum revenue
best_region = ""
highest_revenue = 0

for region, revenue in revenue_per_region.items():
    if revenue > highest_revenue:
        highest_revenue = revenue
        best_region = region

print(f"The region with the highest total revenue is {best_region} with a value of {highest_revenue}.")
```

### Task 2

```python
sales = [1200, 980, 1540, 870, 2010, 1650, 1300, 1100, 950, 1780, 2200, 1430, 1600, 1900]

def summarise_sales(daily_sales):
    total_sales = sum(daily_sales)
    average_sales = round(total_sales / len(daily_sales), 2)
    
    highest_sale = max(daily_sales)
    lowest_sale = min(daily_sales)
    
    # Adding 1 because the index needs to start from 1 instead of 0
    best_day_idx = daily_sales.index(highest_sale) + 1
    worst_day_idx = daily_sales.index(lowest_sale) + 1
    
    above_average_count = 0
    for day in daily_sales:
        if day > average_sales:
            above_average_count = above_average_count + 1
            
    result = {
        "total": total_sales,
        "average": average_sales,
        "best_day": best_day_idx,
        "worst_day": worst_day_idx,
        "above_average_count": above_average_count
    }
    return result

sales_summary = summarise_sales(sales)
print(sales_summary)
```

### Task 3

```python
students = [
    {"name": "Anna", "grade": 85, "attendance": 0.92},
    {"name": "Erik", "grade": 58, "attendance": 0.75},
    {"name": "Mia", "grade": 91, "attendance": 0.98},
    {"name": "Leo", "grade": 44, "attendance": 0.60},
    {"name": "Sara", "grade": 72, "attendance": 0.88},
    {"name": "Tom", "grade": 67, "attendance": 0.70},
    {"name": "Nina", "grade": 95, "attendance": 0.99},
    {"name": "Olli", "grade": 50, "attendance": 0.65},
]

# a) List of passing students
passing = []
for student in students:
    if student["grade"] >= 50:
        passing.append(student["name"])
print(f"Passing students: {passing}\n")

# b) Grade status messages
for student in students:
    if student["grade"] >= 80:
        print(f"{student['name']}: Excellent")
    elif student["grade"] >= 50:
        print(f"{student['name']}: Pass")
    else:
        print(f"{student['name']}: Fail")

# c) Count attendance below 80%
low_attendance_count = 0
for student in students:
    if student["attendance"] < 0.80:
        low_attendance_count = low_attendance_count + 1
print(f"\nStudents with attendance below 0.80: {low_attendance_count}\n")

# d) Highest grade manual loop
highest_grade = 0
top_student_loop = ""

for student in students:
    if student["grade"] > highest_grade:
        highest_grade = student["grade"]
        top_student_loop = student["name"]
print(f"Highest grade (using loop): {top_student_loop} with a grade of {highest_grade}")

# e) Highest grade using max()
all_grades = []
for student in students:
    all_grades.append(student["grade"])

max_grade = max(all_grades)

for student in students:
    if student["grade"] == max_grade:
        print(f"Highest grade (using max): {student['name']} with a grade of {max_grade}")
```

### Task 4

```python
class Portfolio:
    def __init__(self, owner):
        self.owner = owner
        self.holdings = {}

    def add_asset(self, ticker, shares, price_per_share):
        self.holdings[ticker] = {"shares": shares, "price_per_share": price_per_share}

    def total_value(self):
        total = 0
        for data in self.holdings.values():
            total += data["shares"] * data["price_per_share"]
        return total

    def most_valuable(self):
        highest_value = 0
        best_ticker = ""
        
        for ticker, data in self.holdings.items():
            current_value = data["shares"] * data["price_per_share"]
            if current_value > highest_value:
                highest_value = current_value
                best_ticker = ticker
                
        return best_ticker

    def __str__(self):
        return f"Portfolio Owner: {self.owner} | Total Assets: {len(self.holdings)} | Total Value: €{self.total_value():.2f}"

# Testing and methods
my_portfolio = Portfolio("Alice")
my_portfolio.add_asset("AAPL", 50, 150.0)
my_portfolio.add_asset("MSFT", 30, 310.0)
my_portfolio.add_asset("GOOGL", 10, 2800.0)

print(my_portfolio)
print(f"Total Portfolio Value: €{my_portfolio.total_value()}")
print(f"Most Valuable Asset Ticker: {my_portfolio.most_valuable()}")
```

### Task 5

```python
from datetime import datetime

orders = [
    ("2025-01-15", "Laptop"), ("2025-01-22", "Phone"),
    ("2025-02-03", "Laptop"), ("2025-02-14", "Tablet"),
    ("2025-03-01", "Phone"), ("2025-03-10", "Phone"),
    ("2025-03-22", "Laptop"), ("2025-04-05", "Monitor"),
]

parsed_dates = []

# converting string dates to datetime objects
for order in orders:
    date_string = order[0]
    actual_date = datetime.strptime(date_string, "%Y-%m-%d").date()
    parsed_dates.append(actual_date)

earliest_date = min(parsed_dates)
latest_date = max(parsed_dates)

difference = (latest_date - earliest_date).days

print(f"The earliest order was on {earliest_date} and the latest was on {latest_date}.")
print(f"Number of days between them: {difference} days.")
```

```python
import numpy as np

revenue = np.array([
    [210, 230, 190, 250],
    [220, 260, 200, 270],
    [250, 290, 230, 310]
])

# a) row sums
annual_revenue = revenue.sum(axis=1)
print(f"Total annual revenue for 2022, 2023, 2024: {annual_revenue}")

# b) column means
mean_quarterly = revenue.mean(axis=0)
print(f"Mean quarterly revenue for Q1, Q2, Q3, Q4: {mean_quarterly}")

# c) adjusting 2024 row
adjustment_factors = np.array([1.05, 1.03, 1.08, 1.02])
adjusted_2024 = revenue[2] * adjustment_factors
print(f"Adjusted 2024 revenue: {adjusted_2024}")

# d) vAlues > 250
high_revenue = revenue[revenue > 250]
print(f"Quarterly revenues exceeding 250k: {high_revenue}")
```

```python
import pandas as pd

data = {
    "employee": ["Anna", "Erik", "Mia", "Leo", "Sara", "Tom"],
    "department": ["Finance", "Marketing", "Finance", "IT", "Marketing", "IT"],
    "salary": [4500, 4000, 4800, 5200, 3900, 5100],
    "years_exp": [5, 3, 7, 10, 2, 8],
    "performance_score": [4.2, 3.8, 4.5, 4.0, 3.5, 4.9]
}

df = pd.DataFrame(data)

# a) Calculate salary per year of experience
df["salary_per_year_exp"] = (df["salary"] / df["years_exp"]).round(0).astype(int)

# b) Top 3 employees
top_3_employees = df.sort_values(by="performance_score", ascending=False).head(3)
print("Top 3 employees by performance:\n", top_3_employees, "\n")

# c) Finance employees with > 4.0 performance
finance_stars = df.loc[(df["department"] == "Finance") & (df["performance_score"] > 4.0)]
print("Finance employees with performance > 4.0:\n", finance_stars, "\n")

# d) Rename column
df.rename(columns={"years_exp": "experience_years"}, inplace=True)

# e) Select subset using iloc
subset = df.iloc[:4, [2, 4]]
print("Salary and performance for first 4 employees:\n", subset)
```

```python
import numpy as np
import pandas as pd

# Provided data generation script
rng = np.random.default_rng(seed=123)
days = pd.date_range(start="2025-01-06", periods=28, freq="D")
visitors = rng.integers(200, 600, size=28)
conversion_rate = rng.uniform(0.02, 0.08, size=28).round(4)
avg_order_value = rng.normal(50, 10, size=28).round(2)

# a) dataframe
traffic_df = pd.DataFrame({
    "date": days,
    "visitors": visitors,
    "conversion_rate": conversion_rate,
    "avg_order_value": avg_order_value
})
traffic_df.set_index("date", inplace=True)

# b) computing conversion and revenue columns
traffic_df["conversions"] = (traffic_df["visitors"] * traffic_df["conversion_rate"]).round(0).astype(int)
traffic_df["daily_revenue"] = traffic_df["conversions"] * traffic_df["avg_order_value"]

print(traffic_df.head())
```
