```python
import os
import requests
import pandas as pd
from dotenv import load_dotenv

# importing api key
load_dotenv()
api_key = os.getenv("API_KEY")

# 2a) retrieving data
url = "http://api.marketstack.com/v1/eod"
params = {
    "access_key": api_key,
    "symbols": "META,AAPL,AMZN,NFLX,GOOG",
    "date_from": "2026-03-01",
    "date_to": "2026-03-31",
    "limit": 1000
}

# data fetching and transforming to JSON
response = requests.get(url, params=params)
raw_json_data = response.json() 

# 2b) converting to dataframe, the daily records are inside the data key of the JSON
df_faang = pd.DataFrame(raw_json_data['data'])

# converting the date column to actual datetime objects like in task0
df_faang['date'] = pd.to_datetime(df_faang['date'])

print("first 5 rows of FAANG data")
display(df_faang.head())

# 2c) summary stats and returns
print("summary statistics by stock")
display(df_faang.groupby('symbol')['close'].describe())

# date sorting
df_sorted = df_faang.sort_values('date')

# closing prices for stocks in question
first_prices = df_sorted.groupby('symbol')['close'].first()
last_prices = df_sorted.groupby('symbol')['close'].last()

# percentage change formula
returns = ((last_prices - first_prices) / first_prices) * 100
returns_df = returns.reset_index().rename(columns={'close': 'Return_Percentage'})

print("returns in percentages for March 2026")
display(returns_df)

# winners and losers
best = returns_df.loc[returns_df['Return_Percentage'].idxmax()]
worst = returns_df.loc[returns_df['Return_Percentage'].idxmin()]

print(f"\nBest Performing Stock: {best['symbol']} ({best['Return_Percentage']:.2f}%)")
print(f"Worst Performing Stock: {worst['symbol']} ({worst['Return_Percentage']:.2f}%)")
```

```python
import pandas as pd
import sqlite3
import time

# Ensuring the connection
conn = sqlite3.connect('northwind.db')

# --- Task 7a) retrieving full tables into datafarmes 
print("--- Task 7a: Full Tables ---")
df_products_full = pd.read_sql_query("SELECT * FROM Products", conn)
df_categories_full = pd.read_sql_query("SELECT * FROM Categories", conn)

print("Products Table (First 5 rows):")
display(df_products_full.head())
print("\nCategories Table (First 5 rows):")
display(df_categories_full.head())


# Task 7b) perform inner join with pandas
print("\n--- Task 7b: Pandas Inner Join on CategoryID ---")
df_merged_7b = pd.merge(df_products_full, df_categories_full, on='CategoryID')
display(df_merged_7b.head())


# Task 7c) SQL vs Python speeds
print("\n--- Task 7c: Speed Comparison ---")

# 1. Timing the SQL approach
start_sql = time.time()
query_sql = """
SELECT c.CategoryName, SUM(od.UnitPrice * od.Quantity) AS CategoryRevenue
FROM [Order Details] od
JOIN Products p ON od.ProductID = p.ProductID
JOIN Categories c ON p.CategoryID = c.CategoryID
GROUP BY c.CategoryName;
"""
df_sql_result = pd.read_sql_query(query_sql, conn)
end_sql = time.time()
sql_time = end_sql - start_sql

# 2. Timing the Python (pandas) approach
start_python = time.time()

# retrieve raw tables
df_od_raw = pd.read_sql_query("SELECT * FROM [Order Details]", conn)
df_prod_raw = pd.read_sql_query("SELECT * FROM Products", conn)
df_cat_raw = pd.read_sql_query("SELECT * FROM Categories", conn)

# Perform the joins in pandas
merge1 = pd.merge(df_od_raw, df_prod_raw, on='ProductID')
merge2 = pd.merge(merge1, df_cat_raw, on='CategoryID')

# Perform aggregation
merge2['LineTotal'] = merge2['UnitPrice_x'] * merge2['Quantity']
df_python_result = merge2.groupby('CategoryName')['LineTotal'].sum().reset_index(name='CategoryRevenue')

end_python = time.time()
python_time = end_python - start_python

# printing results
print(f"SQL Join Time: {sql_time:.6f} seconds")
print(f"Python Join Time: {python_time:.6f} seconds")

if sql_time < python_time:
    print("\nComment: SQL was faster. Database engines are optimized to process joins and filters internally before returning data.")
else:
    print("\nComment: Python was faster. For very small datasets already loaded in memory, pandas operations can sometimes be quicker.")

# close connection
conn.close()
```

### Task 8

```python
import pandas as pd

# Setup: The Provided Datasets
wide_data = pd.DataFrame({
    "store": ["Helsinki", "Tampere", "Turku"],
    "Q1_sales": [12000, 8500, 9200],
    "Q2_sales": [13500, 9100, 9800],
    "Q3_sales": [11800, 8800, 10100],
    "Q4_sales": [15000, 10200, 11500]
})

long_data = pd.DataFrame({
    "month": ["Jan", "Jan", "Feb", "Feb", "Mar", "Mar"],
    "product": ["Laptop", "Phone", "Laptop", "Phone", "Laptop", "Phone"],
    "units_sold": [45, 120, 50, 110, 55, 130]
})

# Task 8a) Convert wide to long using melt
print("--- Task 8a: Wide to Long ---")
print("BEFORE:")
display(wide_data)

melted_data = pd.melt(wide_data, id_vars=['store'], var_name='quarter', value_name='sales')

print("\nAFTER (Melted):")
display(melted_data)


# Task 8b) Convert long to wide using pivot
print("\n--- Task 8b: Long to Wide ---")
print("BEFORE:")
display(long_data)

pivoted_data = long_data.pivot(index='month', columns='product', values='units_sold').reset_index()

pivoted_data.columns.name = None

print("\nAFTER (Pivoted):")
display(pivoted_data)


# --- Task 8c) Reverting back to original formats
print("Task 8c) Reverting to Originals")

# 1. Revert melted_data back to wide (using pivot)
reverted_wide = melted_data.pivot(index='store', columns='quarter', values='sales').reset_index()
reverted_wide.columns.name = None 

# 2. Revert pivoted_data back to long (using melt)
reverted_long = pd.melt(pivoted_data, id_vars=['month'], var_name='product', value_name='units_sold')

print("Reverted to Original Wide Data:")
display(reverted_wide)

print("\nReverted to Original Long Data:")
display(reverted_long)
```
